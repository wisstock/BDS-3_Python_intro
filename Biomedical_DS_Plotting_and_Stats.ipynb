## **Bar Plots**

## Стовпчасті діаграми

### Create a bar plot from the dataframe showing the mean tip on each day.
---
### Створіть стовпчасту діаграму  що показує середнє значення чайових для кожного дня з датафрейму.

```python
sns.barplot(data=categorical_data, x='day', y='tip')
```

### Create a double-bar plot like above but where mean tips are seperated by sex.
Notice that a key is automatically generated.

### Створіть груповану стовпчасту діаграму де середні чайові розділені за статтю.
Зверніть увагу, що ключ генерується автоматично.


```python
sns.barplot(data=categorical_data, x='day', y='tip',hue='sex')
```

### In both of these bar graphs, the bars represent sample mean and the error bars represent the 95% confidence interval of each mean. However, you can manipulate these defaults using an additional python library, NumPy.
---
На обох цих діаграмах стовці представляють середнє значення вибірки, а лінії помилок — 95% довірчий інтервал кожного середнього значення. Ви можете змінювати параметри діаграми за допомогою додаткової бібліотеки Python, NumPy.

image.png

### Import NumPy.
---
### Імпорт NumPy.

```python
import numpy as np
```

### Create a bar graph like above where each bar represents the sample median instead of sample mean.
---
### Створіть діаграму де кожний стовпець відображає медіану вибірки замість середнього значення.

```python
sns.barplot(x='day', y='tip', data=categorical_data, estimator=np.median)
```

### Notice how the error bars have changed too. They now represent the 95% confidence interval of each sample median instead of the sample mean.
---
### Зверніть увагу, як змінилися також лінії похибки. Тепер вони представляють 95% довірчий інтервал кожної медіани замість  середнього значення.

image.png

### Create a new bar graph where the error bars will now represent a 50% confidence interval.
---
### Створіть нову стовпчату діаграму, де лінії похибки відображатимуть 50% довірчий інтервал.

```python
 sns.barplot(data=categorical_data, x='day', y='tip', ci=50)
```

### Create a new bar graph where the error bars now represent standard deviation (sd)
---
### Створіть нову стовпчасту діаграму, де стовпці похибок тепер представляють стандартне відхилення (sd).
image.png

```python
sns.barplot(data=categorical_data, x='day', y='tip', ci='sd')
```

### Create a bar graph with error bar whiskers that are 20% the width of each category.

### Створіть стовпчасту діаграму з лініями похибок, ширина яких становить 20% від кожної категоріЇ.


```python
sns.barplot(data=categorical_data, x='day', y='tip',capsize=0.2)
```

Create a bar graph with error bar whiskers that are 100% the width of each category.

## **Strip Plots**
---
## **Категорійна точкова діаграма**

Why use strip plots? Consider the situation below:
---
Навіщо використовувати категорійні точкові діаграми? Розглянемо ситуацію нижче:

image.png

Зверніть увагу на те, що коробковий графік не описує розподіл даних (бімодальний) та розмір вибірки.

### Create a strip plot showing all the tips on each day.
---
### Створіть категорійну точкову діаграму з усіма чайовими кожного дня.

```python
sns.stripplot(data=categorical_data, x='day', y='tip')
```

### Notice how the positions of datapoints are randomized each time the code is run.
---
### Зверніть увагу на те, як позиції точок даних розкидуються кожного разу, коли виконується код аби не перекривати одна одну.

```python
sns.stripplot(data=categorical_data, x='day', y='tip')
```

### Create a strip plot like above, but change the points from circles to triangles and adjust their size.
---
### Створіть категорійну точкову діаграму, але змініть точки з кругів на трикутники та налаштуйте їхній розмір.

```python
sns.stripplot(data=categorical_data, x='day', y='tip',marker='^', size=5)
```

### Create a strip plot like the first one, but with the points more spread out.

By default, jitter=1

### Створіть категорійну точкову діаграму, але з більш рознесеними точками.

За замовчуванням jitter=1


```python
sns.stripplot(data=categorical_data, x='day', y='tip',jitter=.3)
```

### Create a strip plot like above, but reduce the alpha value (opacity) of each datapoint.
By default, alpha=1

### Створіть категорійну точкову діаграму, але зменшіть значення alpha (непрозорість) кожної точки даних.
За замовчуванням alpha=1


```python
sns.stripplot(data=categorical_data, x='day', y='tip',jitter=.3,alpha=.5)
```

### Create a double-strip plot where datapoints are labeled by color based on the customer's sex.
---

### Створіть груповану категорійну точкову діаграму, де точки даних позначені кольором залежно від статі клієнта.

```python
sns.stripplot(data=categorical_data, x='day', y='tip',jitter=.3,alpha=.6,hue='sex')
```

### Create a double-strip plot like above, but where alpha is reset and points of different color are seperated.
By default, dodge=False

### Створіть груповану категорійну точкову діаграму, як описано вище, але зі стандартною alpha, а точки іншого кольору розділені.
За замовчуванням dodge=False

```python
sns.stripplot(data=categorical_data, x='day', y='tip',hue='sex',dodge=True)
```
